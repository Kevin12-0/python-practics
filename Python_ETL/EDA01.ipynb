```python
import numpy as np

data = np.random.rand(512,2)
```

```python
data
```

```python
# Generar un scatter plot

px = 1/plt.rcParams['figure.dpi']  # pixel in inches
plt.figure(figsize=(900*px, 700*px))

plt.title('Scatter Plot')
plt.scatter(data[:,0],data[:,1])
plt.show()
```

```python
# Histograma
cm = 1/2.54  # centimeters in inches
plt.figure(figsize=(30*cm, 12*cm))

plt.title('Bar Vertical Plot')
plt.xlabel('Axis "X"')
plt.ylabel('Axis "Y"')
plt.bar(data[20:45,0],data[20:45,1],width=0.05)
plt.show()
```

```python
# Bar Horizontal
cm = 1/2.54  # centimeters in inches
plt.figure(figsize=(20*cm, 20*cm))

plt.title('Bar Horizontal Plot')
plt.xlabel('Axis "X"')
plt.ylabel('Axis "Y"')
plt.barh(data[20:45,0],data[20:45,1],height=.05)
plt.show()
```

```python
# Bar con grupos

A = [2,22,44,88]
B = [1,11,33,66]

X = range(4)

# Bar Vertical
cm = 1/2.54  # centimeters in inches
plt.figure(figsize=(20*cm, 20*cm))

plt.title('Bar Vertical Group Plot')
plt.xlabel('Axis "X"')
plt.ylabel('Axis "Y"')
plt.bar(X,A,color='g')
plt.bar(X,B,color='r',bottom=A)
plt.legend(["green","red"], loc="upper center")
plt.show()
```

```python
# Gr√°fico de Pastel

A = [2,22,44,88]
cm = 1/2.54  # centimeters in inches
plt.figure(figsize=(20*cm, 20*cm))

plt.pie(A)
plt.legend(["blue","yellow","green","red"], loc="lower right")
plt.show()
```

```python
# Ejemplo 1 mas completo:

import matplotlib.pyplot as plt
import numpy as np
 
city=['Delhi','Beijing','Washington','Tokyo','Moscow']
Gender=['Male','Female']
pos = np.arange(len(city))
Happiness_Index_Male=[60,40,70,65,85]
Happiness_Index_Female=[30,60,70,55,75]

cm = 1/2.54  # centimeters in inches
plt.figure(figsize=(30*cm, 20*cm)) 
plt.bar(pos,Happiness_Index_Male,color='blue',edgecolor='black')
plt.bar(pos,Happiness_Index_Female,color='pink',edgecolor='black',bottom=Happiness_Index_Male)
plt.xticks(pos, city)
plt.xlabel('City', fontsize=16)
plt.ylabel('Happiness_Index', fontsize=16)
plt.title('Stacked Barchart - Happiness index across cities',fontsize=18)
plt.legend(Gender,loc=2)
plt.show()
```

```python
#Ejemplo 2:

import matplotlib.pyplot as plt
 
values = [60, 80, 90, 55, 10, 30]
colors = ['b', 'g', 'r', 'c', 'm', 'y']
labels = ['US', 'UK', 'India', 'Germany', 'Australia', 'South Korea']
explodex = (0.2, 0, 0, 0, 0, 0)

cm = 1/2.54  # centimeters in inches
plt.figure(figsize=(30*cm, 20*cm)) 
plt.pie(values, colors=colors, labels=labels, explode = explodex, autopct='%1.1f%%',counterclock=True, shadow=True)
plt.title('Population Density Index')
plt.show()
    
    
```

```python

import seaborn as sns

# Cargar DataSet de Biblioteca
df = sns.load_dataset('penguins')

df
```

```python
# Generar una l√≠nea de Regresi√≥n Simple lmplot

# Aplicar Tema por defecto - Cuadros
sns.set_theme()

rl01 = sns.lmplot(
                 data=df, 
                 x="bill_length_mm", 
                 y="flipper_length_mm")
```

```python
# Propiedad hue

sns.set_theme()

# Generar Regresi√≥n L√≠neal Simple

rl02 = sns.lmplot(data=df, x="bill_length_mm", y="flipper_length_mm",hue="sex")
```

```python
# Propiedad col

sns.set_theme()

rl03 = sns.lmplot(data=df, x="bill_length_mm", y="flipper_length_mm", col="species", hue="sex")
```

```python
# Propiedad FacetGrid

hist01 = sns.FacetGrid(df, col="species", height=3.5, aspect=.65)
hist02 = hist01.map_dataframe(sns.histplot, x="bill_length_mm", kde=True, hue = "sex")
```

```python
# Propiedad element 

hist03 = sns.histplot(df, x="flipper_length_mm", hue="species", element="poly")
```
