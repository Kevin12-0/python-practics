```python
# Genera el grupo como Tabla (as_index = False) y ordena

#df_sort01 = pd.DataFrame(dfuser.groupby(['Region'],as_index=False)['Sales'].mean())

df_sort01 = pd.DataFrame(dfuser.groupby(['Region'],as_index=True)['Sales'].sum())
                         
df_sort01 = df_sort01.sort_values('Sales',ascending=True)

df_sort01
```

```python
# Genera el grupo como Tabla Pivotante (as_index = True) y ordena

df_sort02 = pd.DataFrame(dfuser.groupby(['Region','Category','SubCategory'],as_index=False)[['Profit','Sales']].sum())
df_sort02 = df_sort02.sort_values(['Profit','Sales'],ascending= True)
df_sort02
```

```python
# Función AGG

round(dfuser.agg({'Sales'  :['min','max','mean','median','std','var'],
            'Profit' :['mean','median']
                 }),2)
```

```python
# Similar con Describe

round(dfuser.describe(include='all'),2)
```

```python
# Combinar Group by con Agg

dfuser.groupby(['Region','Category','SubCategory'],as_index=True).agg({'Sales':['mean','median','std','var'],
                                                                       'Profit': ['mean','median','std','var']
                                                                      })
```

```python
# Group by con la función Aggregate - otra alternativa

import numpy as np

df_sort03 = pd.DataFrame(round(dfuser.groupby(['Country'],as_index=True)
            ['Profit'].aggregate([min,np.std,np.mean,max]),2))
df_sort03
```

```python
# Operaciones Join

df1 = pd.DataFrame({'employee': ['Bob', 'Jake'],
                       'group': ['Accounting', 'Engineering']})

df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake'],
                    'hire_date': [2004, 2008,2000]})
```

```python
df3 = pd.merge(df1,df2)
df3
```

```python
df4 = pd.merge(df1, df2, on='employee')
df4
```

```python
# Operaciones Join modificadas

df1 = pd.DataFrame({'employee': ['Bob', 'Jake','larson'],
                       'group': ['Accounting', 'Engineering','Data Engineer']})

df2 = pd.DataFrame({'empleado': ['Lisa', 'Bob'],
                    'hire_date': [2004, 2008],
                    'group':['Human Reosurces','Architect']
                   })
```

```python
pd.merge(df2, df1, left_on="empleado", right_on="employee")
```

```python
df03 = pd.merge(df1, df2, left_on="employee", right_on="empleado").drop('employee', axis=1)

df03[['hire_date']]
```

```python
pd.merge(df1, df2, how='inner',left_on="employee", right_on="empleado")
```

```python
pd.merge(df1, df2, how='left',left_on="employee", right_on="empleado")
```

```python
pd.merge(df1, df2, how='right',left_on="employee", right_on="empleado")
```

```python
pd.merge(df1, df2, how='outer',left_on='employee',right_on='empleado')
```
